```python
data = pd.read_csv("week2.csv")
data['Date'] = data['Date'].astype(dtype = 'datetime64')
data.set_index(['Date'],inplace = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import warnings
warnings.filterwarnings('ignore')
%matplotlib notebook
```

```python
plt.figure()
plt.plot(data.index,data['Close Price'])
plt.ylabel('Close Price') #it provides label on y axis
plt.xlabel('Date') #it provides label on x axis
plt.xticks(rotation = 45)
plt.title('Closing Price on each day')#it provides title on the graph
plt.grid(True,axis = 'both')
plt.show()
```

```python
plt.figure()
plt.stem(data.index,data['Day_Perc_Change'])
plt.ylabel('Percentage Change') #it provides label on y axis
plt.xlabel('Date') #it provides label on x axis
plt.xticks(rotation = 45)
plt.title('Percentage Change in Price')#it provides title on the graph
plt.grid(True,axis = 'both')
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(2,sharex=True)
fig.suptitle('Vertically stacked subplots')#it provides title to the graph
ax1.grid(True,axis='both')
ax1.plot(data.index,data['Total Traded Quantity'])
ax2.grid(True,axis='both')
plt.title('Percentage change in Price')#it provides title to the graph
ax2.stem(data.index,data['Day_Perc_Change'])
ax2.plot(data.index,data['Total Traded Quantity'].pct_change(),color = 'black')
plt.xticks(rotation = 45)
plt.show( )
```

```python
"""Volume is greater than the daily percentage change.the relationship between volume and daily percentage change is directly proportional. if volume increases, the daily percentage change also increases."""
```

```python
data['Trend Occurence'] = np.ones(len(data))
data2 = data[['Trend Occurence','Trend']].groupby('Trend').sum()/len(data)*100
sizes = data2['Trend Occurence']
labels = data2.index
plt.figure()
plt.pie(sizes,labels =labels,autopct='%1.2f%%',startangle=90)
plt.axis('equal')
plt.show()
data.drop(['Trend Occurence'],axis =1,inplace = True)
del(data2)
```

```python
fig1,(ax2,ax3) = plt.subplots(2,sharex = True)
fig1.suptitle("Mean and Median values of volume in each Trends")
ax2.grid(True , axis = 'both')
ax3.grid(True , axis = 'both')
data2 = data[['Trend','Total Traded Quantity']].groupby('Trend').mean()
ax2.bar(data2.index,data2['Total Traded Quantity'],label = 'mean')
data2 = data[['Trend','Total Traded Quantity']].groupby('Trend').median()
ax3.bar(data2.index,data2['Total Traded Quantity'],color = 'red',label = "median")
ax2.legend()
ax3.legend()
plt.xlabel("Trend")
ax2.set(ylabel = "mean Values")
ax3.set(ylabel = "median Values")
plt.show()
del(data2)
```

```python
df1 = pd.read_csv('Small_Cap/IDFC.csv') #imported IDFC.CSV
df1 = df1[df1.Series == 'EQ']
df1['Date'] = df1['Date'].astype('datetime64')
df1.set_index(['Date'],inplace = True)
```

```python
df2 = pd.read_csv('Small_Cap/FORTIS.csv') #imported FORTIS.CSV
df2 = df2[df2.Series == 'EQ']
df2['Date'] = df2['Date'].astype('datetime64')
df2.set_index(['Date'],inplace = True)
```

```python
df3 = pd.read_csv('Small_Cap/NCC.csv') #imported NCC.CSV
df3 = df3[df3.Series == 'EQ']
df3['Date'] = df3['Date'].astype('datetime64')
df3.set_index(['Date'],inplace = True)
```

```python
df4 = pd.read_csv('Small_Cap/PVR.csv') #imported PVR.CSV
df4 = df4[df4.Series == 'EQ']
df4['Date'] = df4['Date'].astype('datetime64')
df4.set_index(['Date'],inplace = True)
```

```python
df5 = pd.read_csv('Small_Cap/RCOM.csv') #imported RCOM.CSV
df5 = df5[df5.Series == 'EQ']
df5['Date'] = df5['Date'].astype('datetime64')
df5.set_index(['Date'],inplace = True)
```

```python
Close_price_df = pd.DataFrame({'IDFC':df1['Close Price'],'FORTIS':df2['Close Price'],'NCC':df3['Close Price'],'PVR':df4['Close Price'],'RCOM':df5['Close Price']})
```

```python
Percent_Close_price_df = Close_price_df.pct_change()*100
Percent_Close_price_df.dropna(inplace = True)
```

```python
corr = Percent_Close_price_df.corr()
mask = np.zeros_like(corr,dtype=np.bool)
mask[np.triu_indices_from(mask)]=True
f,ax = plt.subplots()
cmap = sns.diverging_palette(220,10,as_cmap=True)
sns.heatmap(corr,mask=mask,cmap=cmap,vmax=.3,center=0,square=True,linewidths=.5,cbar_kws={"shrink":.5})
```

```python
df = Percent_Close_price_df['FORTIS'].rolling(7).mean()
vol = Percent_Close_price_df['FORTIS'].rolling(7).std()*np.sqrt(7)
plt.figure()
vol.plot(label = "Varience")
df.plot(label = "Average",color = "red")
plt.legend()
plt.xlabel('Dates')#it provides label on x axis
plt.title("Rolling change in average and variance of percentage change of fortis")
plt.ylabel("Change")
plt.xticks(rotation = 45)
plt.show()
```

```python
print("The Standard Deviation ="+str(Percent_Close_price_df['FORTIS'].std()))
```

```python
df_Nifty = pd.read_csv("Nifty50/NIfty50.csv")
df_Nifty['Date'] = df_Nifty['Date'].astype('datetime64')
df_Nifty.set_index('Date',inplace = True)
df_Nifty_ClsPrc = df_Nifty['Close'].pct_change()*100
vol2 = df_Nifty_ClsPrc.rolling(7).std() * np.sqrt(7)
plt.figure()
vol2.plot(label = "Nifty")
vol.plot(label = "Fortis")
plt.legend()
plt.show()
```

```python
v = df2['Close Price'].rolling(14).mean()
vu = v + 2*df2['Close Price'].rolling(14).std()
v1 =  v - 2*df2['Close Price'].rolling(14).std()
plt.figure()
plt.title("Bollinger Bands of Fortis")
plt.ylabel("Prices")
vu.plot(label = "+2 Std Dev",alpha =0.7)
v.plot(label = 'Average',alpha =0.7)
vu.plot(label = "-2 Std Dev",alpha =0.7)
df2['Average Price'].plot(label = "Daily Average",color ="black")
plt.legend()
plt.show()
```
