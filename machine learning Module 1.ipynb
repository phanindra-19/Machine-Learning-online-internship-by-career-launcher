```python
data1['Month'] = pd.DatetimeIndex(data1['Date']).month
data1['Year'] = pd.DatetimeIndex(data1['Date']).year
data1['VWAP'] = (data1['Close Price'] * data1['Total Traded Quantity']).cumsum()/data1['Total Traded Quantity'].fillna(0).cumsum()
```

```python
data_vwap = data1 [['Month','Year','VWAP']]
group = data_vwap.groupby(['Month','Year'])
group.first()
```

```python
def avg_price(N):
    return (data1['Average Price'].tail(N).sum())/N
print("Average prices for last  N days are as follows:")
print("Last 1 week",avg_price(5))
print("Last 2 weeks",avg_price(10))
print("Last 1 month",avg_price(20))
print("Last 3 months",avg_price(60))
print("Last 6 months",avg_price(120))
print("Last 1 year",avg_price(240))
print("Profit/Loss % for N days are as follows:")
def prof_loss(N):
    difference = (data1['Close Price'].tail(N).iloc[N-1] - data1['Close Price'].tail(N).iloc[0])
    if difference < 0:
        loss = -(difference)
        loss_percen = (loss/data1['Close Price'].tail(N).iloc[N-1])*100
        return loss_percen
    if difference > 0:
        profit = difference
        profit_percen = (profit/data1['Close Price'].tail(N).iloc[N-1])*100
        return profit_percen
print("Loss/Profit percentage for the last N days are as follows:")
print("Last 1 week",prof_loss(5))
print("Last 2 weeks",prof_loss(10))
print("Last 1 month",prof_loss(20))
print("Last 3 months",prof_loss(60))
print("Last 6 months",prof_loss(120))
print("Last 1 year",prof_loss(240))

```

```python
data1['Day_Perc_Change'] = data1['Close Price'].pct_change().fillna(0)
```

```python
data1
```

```python
data1['Trend']='slight or no change'
data1
```

```python
if((data1['Day_Perc_Change'] >= -0.5) & (data1['Day_Perc_Change'] <= 0.5)).all():
    data1['Trend'] = 'Slight change or No change'
if((data1['Day_Perc_Change'] >= 0.5) & (data1['Day_Perc_Change'] <= 1)).all():
    data1['Trend'] = 'Slight positive'
if((data1['Day_Perc_Change'] >= -0.5) & (data1['Day_Perc_Change'] <= - 1)).all():
    data1['Trend'] = 'Slight negative'
if((data1['Day_Perc_Change'] >= 1) & (data1['Day_Perc_Change'] <= 3)).all():
    data1['Trend'] = 'positive'
if((data1['Day_Perc_Change'] >= -1) & (data1['Day_Perc_Change'] <= -3)).all():
    data1['Trend'] = 'negative'
if((data1['Day_Perc_Change'] >= 3) & (data1['Day_Perc_Change'] <= 7)).all():
    data1['Trend'] = 'among top gainers'
if((data1['Day_Perc_Change'] >= -3) & (data1['Day_Perc_Change'] <= -7)).all():
    data1['Trend'] = 'among top lossers'
if(data1['Day_Perc_Change'] > 7).all():
    data1['Trend'] = 'Bull Run'
if(data1['Day_Perc_Change'] >= -7).all():
    data1['Trend'] = 'Bear Drop'
```
