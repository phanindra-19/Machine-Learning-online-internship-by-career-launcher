```python
import pandas as pd
gold_data = pd.read_csv('GOLD.csv')
gold_data
```

```python
gold_data.set_index('Date',inplace=True)
gold_data
```

```python
gold_without_nan = gold_data.dropna()
gold_without_nan
```

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#dataset
y = np.array(gold_without_nan["Pred"])
x = np.array(gold_without_nan["new"])
x = x.reshape(-1,1)
y = y.reshape(-1,1)
# sckit-learn implementation
# Model initialization

regression_model = LinearRegression()
# Fit the data(train the model)
regression_model.fit(x, y)
# Predict
y_predicted = regression_model.predict(x)

# model evaluation
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

# printing values
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)

# plotting values

# data points
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='r')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

'''
Download monthly prices of RCOM and NIFTY 50 for Time period: 1-Jan-2014--12-Jan-2017
'''
RCOM = pd.read_csv('RCOM.csv', parse_dates=True, index_col='Date',)
nifty50 = pd.read_csv('Nifty50/Nifty50.csv', parse_dates=True, index_col='Date')

# joining the closing prices of the two datasets 
monthly_prices = pd.concat([RCOM['Close Price'], nifty50['Close']], axis=1)
monthly_prices.columns = ['RCOM', 'NIFTY50']

# check the head of the dataframe
print(monthly_prices.head())

# calculate monthly returns
monthly_returns = monthly_prices.pct_change(1)
clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row
print(clean_monthly_returns.head())
```

```python
# split dependent and independent variable
X = clean_monthly_returns['RCOM']
y = clean_monthly_returns['NIFTY50']

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```

```python
print("Daily beta value for the past 3 months for the stock RCOM is -0.0188 which is less than 1 and negative.This means it has a negative relationship between variables.The increase in the first variable will cause the decrease in the second variable and vice versa. Hence it is less volatile than the benchmark. Therefore,there exists negative regression.")
```
